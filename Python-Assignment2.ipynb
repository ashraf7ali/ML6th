1. Create the below pattern using nested for loop in Python.

```python
n = 6
for i in range(1,n):
    for j in range(i):
        print('*', end = ' ')
    print('')

for k in range(4,0,-1):
    for l in range(k):
        print('*', end = ' ')
    print('')
```
